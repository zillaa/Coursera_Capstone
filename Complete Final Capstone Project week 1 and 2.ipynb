#### We can see that the there is very slight uptick in the prices across vancouver for the year 2019, Lets wrangle and clean up the data as some of the data is missing from the source and lets also remove unwanted columns, for clarity.

```python
van_df.columns
```

```python
v_df = van_df[['PROPERTY_POSTAL_CODE','CURRENT_LAND_VALUE','PREVIOUS_LAND_VALUE']]
v_df
```

```python
#lets assign meaningful names to the columns.
v_df.columns = ['postal code','Land Value 2019','Land Value 2018']
v_df.tail()
```

```python
#now lets keep all the areas where the land value is more than 2018.
in_df = v_df[v_df['Land Value 2019']> v_df['Land Value 2018']]
in_df.tail()
```

```python
in_df.shape
```

```python
#now lets find the areas that are in the price range of 3 million and 3.5 million.
df = in_df[(in_df['Land Value 2019'] > 3000000) & (in_df['Land Value 2019'] < 3500000)]
df
```

```python
df.shape
```

```python
#now lets drop same postal codes
df = df.drop_duplicates(subset='postal code', keep='first')
df
```

```python
df.shape
```

```python
df.index = range(9)
df
```

```python
# Setup API URL to explore venues near by
LIMIT = 100
radius = 500
url = 'https://api.foursquare.com/v2/venues/explore?client_id={}&client_secret={}&ll={},{}&v={}&radius={}&limit={}'.format(CLIENT_ID, CLIENT_SECRET, pc_latitude, pc_longitude, VERSION, radius, LIMIT)
neighborhood_json = requests.get(url).json()

# function to extract category of the venue
def get_category_type(row):
    try:
        categories_list = row['categories']
    except:
        categories_list = row['venue.categories']
        
    if len(categories_list) == 0:
        return None
    else:
        return categories_list[0]['name']
    
venues = neighborhood_json['response']['groups'][0]['items']
    
nearby_venues = json_normalize(venues) # flatten JSON

# filter columns
filtered_columns = ['venue.name', 'venue.categories', 'venue.location.lat', 'venue.location.lng']
nearby_venues =nearby_venues.loc[:, filtered_columns]

# filter the category for each row
nearby_venues['venue.categories'] = nearby_venues.apply(get_category_type, axis=1)

# clean columns
nearby_venues.columns = [col.split(".")[-1] for col in nearby_venues.columns]

nearby_venues
```
